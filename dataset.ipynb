```python
from PIL import Image, ImageDraw
import random
```

```python
# рисуем круг
def draw_circle(draw: ImageDraw, x: int, y: int, color: tuple[int, int, int]):
    draw.point([(x-1, y), (x+1, y), (x, y-1), (x, y+1)], fill=color)
    return

# рисуем квадрат


def draw_box(draw: ImageDraw, x: int, y: int, color: tuple[int, int, int]):
    draw.line([(x-1, y-1), (x+1, y-1), (x+1, y+1),
              (x-1, y+1), (x-1, y-1)], fill=color)
    return

# рисуем плюс


def draw_plus(draw: ImageDraw, x: int, y: int, color: tuple[int, int, int]):
    draw.line([(x-1, y), (x+1, y)], fill=color)
    draw.line([(x, y-1), (x, y+1)], fill=color)
    return

# рисуем крест


def draw_cross(draw: ImageDraw, x: int, y: int, color: tuple[int, int, int]):
    draw.line([(x-1, y-1), (x+1, y+1)], fill=color)
    draw.line([(x+1, y-1), (x-1, y+1)], fill=color)
    return

# рисуем черту


def draw_slash(draw: ImageDraw, x: int, y: int, color: tuple[int, int, int]):
    draw.line([(x+1, y-1), (x-1, y+1)], fill=color)
    return


def draw_square(draw: ImageDraw, x: int, y: int, color: tuple[int, int, int]):
    for i in range(-1, 2):
        for j in range(-1, 2):
            draw.point([(x + i, y + j)], fill=color)


# вероятность вызова draw_slash увеличена в 3 раза
drawing_functions = [
    draw_circle, draw_box, draw_plus, draw_cross, draw_slash,
    draw_slash, draw_slash
]

drawing_colors = [
    (255, 0, 0),
    (0, 255, 0),
    (0, 0, 255),
]

# Генерируем датасет
for image_index in range(0, 1000):
    img = Image.new('RGB', (48, 48))
    draw = ImageDraw.Draw(img)

    img_etalon = Image.new('RGB', (48, 48))
    draw_etalon = ImageDraw.Draw(img_etalon)

    for i in range(1, 48, 3):
        for j in range(1, 48, 3):
            random_function = random.choice(drawing_functions)
            random_color = random.choice(drawing_colors)

            if random_function == draw_slash:
                random_function(draw, i, j, (255, 255, 0))
                draw_square(draw_etalon, i, j, (255, 255, 255))
                continue

            random_function(draw, i, j, random_color)
            random_function(draw_etalon, i, j, random_color)

    img.save(f'data/{image_index}_test.png', resample=Image.NEAREST)
    img_etalon.save(f'data/{image_index}_etalon.png', resample=Image.NEAREST)

    img.close()
    img_etalon.close()
```
